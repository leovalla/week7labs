## Sorting and Ordering data 

```python
# Select `human-bot` teams that have a `bot_members_count` value greater than and equal to 2.
gh_teams[(gh_teams.Team_type == 'human-bot') & (gh_teams.bot_members_count >= 2)] 
```

```python
# Find the `human` teams that are `Large` and have a `human_gini` value greater than and equal to 0.75.
gh_teams[(gh_teams.Team_type == 'human') &
(gh_teams.Team_size_class == 'Large') &
(gh_teams.human_gini >= 0.75)] 
```

```python
# How many teams are in the `Small` or `Large` category?
teams_number = len(gh_teams[(gh_teams.Team_size_class == 'Large') | (gh_teams.Team_size_class == 'Small') ])
print(f"There are {teams_number} small or lager teams")
```

```python
# How many teams are in the `Small` or `Large` category with a `human_gini` value less than and equal to 0.20?
teams_number_2 = len(gh_teams[((gh_teams.Team_size_class == 'Large') | (gh_teams.Team_size_class == 'Small')) &
                     (gh_teams.human_gini <= 0.2)])
print(f"There are {teams_number_2} small or lager teams a with a human_gini value less than and equal to 0.20")
```

```python
# How many `human-bot` teams are in the `Medium` category?
teams_number_3 = len(gh_teams[(gh_teams.Team_type == 'human-bot') &
(gh_teams.Team_size_class == 'Medium')])

print(f"There are {teams_number_3} human-bot teams whit a Medium category")
```

```python
# Create a subsample of 50% of your data.
gh_teams_subsample = gh_teams.sample(frac = 0.5, replace = False)
len(gh_teams_subsample.index)
```

```python
# Create samples for a 8-fold cross validation test
kf = KFold(n_splits = 8)  # K-Fold cross-validator with 8 folds 

for train, test in kf.split(gh_teams):
    print("%s %s" % (train, test))
```
